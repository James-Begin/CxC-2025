```python
import pandas as pd

df = pd.read_excel('SAP Datasets.xlsx')

year_cols = [str(year) for year in range(2000, 2024)]

keep_cols = ['Country Name', 'Indicator Name'] + year_cols

df_clean = df[keep_cols]

print(df_clean)
```

```python
indicator_validity = df_clean.groupby('Indicator Name')[year_cols].apply(
    lambda x: x.notna().sum().sum() / (x.shape[0] * len(year_cols))
)

valid_indicators = indicator_validity[indicator_validity > 0.5].index.tolist()

df_final = df_clean[df_clean['Indicator Name'].isin(valid_indicators)]

print(df_final)
```

```python
unique_values = df_final['Indicator Name'].unique()
print(unique_values)

```

```python
import pandas as pd

selected_indicators = [
    'Access to electricity (% of population)',
    'Access to clean fuels and technologies for cooking (% of population)',
    'Compulsory education, duration (years)',
    'Wage and salaried workers, total (% of total employment) (modeled ILO estimate)',
    'Adjusted net national income per capita (current US$)',
    'Age dependency ratio (% of working-age population)',
    'Adolescent fertility rate (births per 1,000 women ages 15-19)',
    'People using at least basic sanitation services (% of population)'
]

indicator_weights = {
    'Access to electricity (% of population)': 0.1515,
    'Access to clean fuels and technologies for cooking (% of population)': 0.1492,
    'Compulsory education, duration (years)': 0.0762,
    'Wage and salaried workers, total (% of total employment) (modeled ILO estimate)': 0.1411,
    'Adjusted net national income per capita (current US$)': 0.1192,
    'Age dependency ratio (% of working-age population)': 0.1169,
    'Adolescent fertility rate (births per 1,000 women ages 15-19)': 0.0908,
    'People using at least basic sanitation services (% of population)': 0.1551
}



indicators_to_invert = [
'Age dependency ratio (% of working-age population)',
'Adolescent fertility rate (births per 1,000 women ages 15-19)'
]
filtered_df = df_final[df_final['Indicator Name'].isin(selected_indicators)]

df_long = filtered_df.melt(id_vars=['Country Name', 'Indicator Name'],
                           value_vars=year_cols,
                           var_name='Year',
                           value_name='Value')

df_pivot = df_long.pivot_table(index=['Country Name', 'Year'],
                               columns='Indicator Name',
                               values='Value').reset_index()

for indicator in selected_indicators:
    if indicator in df_pivot.columns:
        country_means = df_pivot.groupby('Country Name')[indicator].transform('mean')

        df_pivot[indicator] = df_pivot[indicator].fillna(country_means)

print(df_pivot.head())

def invert_value(x):
  return 1 / x if x != 0 else 0

for indicator in indicators_to_invert:
  if indicator in df_pivot.columns:
    df_pivot[indicator] = df_pivot[indicator].apply(invert_value)

for indicator in selected_indicators:
    if indicator in df_pivot.columns:
        df_pivot[indicator] = df_pivot[indicator].fillna(0)

mpi_values = []
for index, row in df_pivot.iterrows():
    mpi_value = sum(row[indicator] * indicator_weights[indicator] for indicator in selected_indicators if indicator in row)
    mpi_values.append(mpi_value)

df_pivot['MPI'] = mpi_values

print(df_pivot[['Country Name', 'Year', 'MPI']].head())
df_pivot.to_excel('mpi_per_country_per_year.xlsx', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

selected_indicators = [
    'Access to electricity (% of population)',
    'Access to clean fuels and technologies for cooking (% of population)',
    'Compulsory education, duration (years)',
    'Wage and salaried workers, total (% of total employment) (modeled ILO estimate)',
    'Adjusted net national income per capita (current US$)',
    'Age dependency ratio (% of working-age population)',
    'Adolescent fertility rate (births per 1,000 women ages 15-19)',
    'People using at least basic sanitation services (% of population)'
]

shortened_names = {
    'Access to electricity (% of population)': 'Electricity',
    'Access to clean fuels and technologies for cooking (% of population)': 'Clean Fuels',
    'Compulsory education, duration (years)': 'Comp Edu.',
    'Wage and salaried workers, total (% of total employment) (modeled ILO estimate)': 'Wage Workers',
    'Adjusted net national income per capita (current US$)': 'Adj. Income',
    'Age dependency ratio (% of working-age population)': 'Age Ratio',
    'Adolescent fertility rate (births per 1,000 women ages 15-19)': 'Adolescent Fertility',
    'People using at least basic sanitation services (% of population)': 'Sanitation',
    'MPI': 'MPI'
}

df_corr = df_pivot[selected_indicators + ['MPI']]

correlation_matrix = df_corr.corr()
correlation_matrix.rename(columns=shortened_names, index=shortened_names, inplace=True)
print("Correlation Matrix:")
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Each Indicator and MPI")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

selected_indicators = [
    'Access to electricity (% of population)',
    'Access to clean fuels and technologies for cooking (% of population)',
    'Compulsory education, duration (years)',
    'Wage and salaried workers, total (% of total employment) (modeled ILO estimate)',
    'Adjusted net national income per capita (current US$)',
    'Age dependency ratio (% of working-age population)',
    'Adolescent fertility rate (births per 1,000 women ages 15-19)',
    'People using at least basic sanitation services (% of population)'
]

X = df_pivot[selected_indicators]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=len(selected_indicators))
pca.fit(X_scaled)

explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio for each component:")
for i, var in enumerate(explained_variance):
    print(f"PC{i+1}: {var:.4f}")

loadings = pd.DataFrame(
    pca.components_.T,
    index=selected_indicators,
    columns=[f'PC{i+1}' for i in range(len(selected_indicators))]
)

print("\nPCA Loadings:")
print(loadings)
grouped_loadings = loadings.rename(index=shortened_names)
plt.figure(figsize=(10, 6))
plt.bar(grouped_loadings.index, grouped_loadings['PC1'])
plt.xticks(rotation=45, ha='right')
plt.ylabel('Loading Value')
plt.title('Feature Loadings on MPI')
plt.tight_layout()
plt.show()
```

```python
pc1_loadings = np.abs(loadings['PC1'])
weights = pc1_loadings / pc1_loadings.sum()

print("Weights Based on PCA Loadings (PC1):")
for name, weight in zip(loadings.index, weights):
    print(f"{name}: {weight:.4f}")
```
